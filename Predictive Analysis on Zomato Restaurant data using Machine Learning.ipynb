```python
countrycode=pd.read_excel('Country-Code.xlsx',engine='openpyxl')
```

```python
countrycode
```

```python
resturants = resturants.merge(countrycode, on='Country Code')
```

```python
resturants.head()
```

```python
indiaRes=resturants[resturants['Country']=='India']
singaporeRes=resturants[resturants['Country']=='Singapore']
AusRes=resturants[resturants['Country']=='Australia']
BrazRes=resturants[resturants['Country']=='Brazil']
CanRes=resturants[resturants['Country']=='Canada']
UAERes=resturants[resturants['Country']=='UAE']
UKRes=resturants[resturants['Country']=='United Kingdom']
USRes=resturants[resturants['Country']=='United States']
```

```python
indiaRes.head()
```

```python
singaporeRes.head()
```

```python
indiaRes.City.value_counts()
```

```python
singaporeRes.City.value_counts()
```

```python
DelhiTopRes=indiaRes[(indiaRes['City']=='New Delhi') & (indiaRes['Aggregate rating']>4.5)][['Aggregate rating','Restaurant Name','Votes']].sort_values(ascending=False, by ='Votes')
```

```python
DelhiTopRes["Aggregate rating"] = DelhiTopRes["Aggregate rating"].round(1)
```

```python
DelhiTopRes['Aggregate rating']
```

```python
DelhiTopRes.style.apply(lambda x:['background: green' if x.name=='Votes' else 'background: lightsteelblue' for i in x])
```

```python
singaporeRes.City.value_counts()
```

```python
singaporeTopRes=singaporeRes[(singaporeRes['City']=='Singapore') & (singaporeRes['Aggregate rating']>3.5)][['Aggregate rating','Restaurant Name','Votes']].sort_values(ascending=False, by ='Votes')
```

```python
singaporeTopRes.style.apply(lambda x:['background: pink' if x.name=='Votes' else 'background:lightsteelblue' for i in x])
```

### finding out relnship between the rating and votes
